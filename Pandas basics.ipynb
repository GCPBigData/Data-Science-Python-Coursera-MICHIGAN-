### Use of builtin methods vs Series Iteration

#### sum

```python
s = pd.Series(np.random.randint(0,1000,10000))
```

```python
%%timeit -n 100
total = 0
for item in s:
    total += item
```

```python
%%timeit -n 100
SUM = s.sum()
```

#### "+="

```python
%%timeit -n 3
s = pd.Series(np.random.randint(0,1000,10000))
for label, value in s.iteritems():
    s.loc[label] = value+2
```

```python
%%timeit -n 3
s = pd.Series(np.random.randint(0,1000,10000))
s += 2
```
